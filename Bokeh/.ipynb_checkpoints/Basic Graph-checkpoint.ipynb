```python
# making a basic bokeh line graph

#Importing bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

#prepare some data
X = [1,2,3,4,5]
Y = [6,7,8,9,10]
A = [12, 13, 14, 15]
B = [1, 3, 4, 6]

#prepare the output file
output_file("line.html")

#create a figure object
f = figure()

#create line plot
f.line(X, Y)

#Show your graph
show(f)

```

```python
output_file("circle.html")

c = figure()
c.circle(X,Y)
show(c)
```

```python
output_file("triangle.html")

A = [12, 13, 14, 15]
B = [1, 3, 4, 6]

t = figure()
t.triangle(A,B)
show(t)
```

```python
# plotting education data from pythonHow

#Importing bokeh and pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

#prepare some data
df = pandas.read_excel('http://pythonhow.com/data/verlegenhuken.xlsx')
X = df['Temperature']
Y = df['Pressure']


#prepare the output file
output_file("tempAndPressure.html")

#create a figure object
f = figure(plot_width=500,plot_height=400,tools='pan')

#Add properties
f.title.text="Temperature and Air Pressure"

f.title.text_color="Gray"

f.title.text_font="times"

f.title.text_font_style="bold"

f.xaxis.minor_tick_line_color=None

f.yaxis.minor_tick_line_color=None

f.xaxis.axis_label="Temperature (â„ƒ)"

f.yaxis.axis_label="Pressure (hPa)"  

#create plot
f.circle(X, Y, size=0.5)

#Show your graph
show(f)

```

```python
from bokeh.plotting import figure, output_file, show

p = figure(plot_width=500, plot_height=400, tools = 'pan, reset')

p.title.text = "Earthquakes"

p.title.text_color = "Orange"

p.title.text_font = "times"

p.title.text_font_style = "italic"

p.yaxis.minor_tick_line_color = "Yellow"

p.xaxis.axis_label = "Times"

p.yaxis.axis_label = "Value"

p.circle([1,2,3,4,5], [5,6,5,5,3], size = [i*2 for i in [8,12,14,15,20]], color="red", alpha=0.5)

output_file("Scatter_plotting.html")

show(p)
```
